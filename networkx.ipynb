```python
import networkx as nx
import pandas as pd
import numpy as np
node = list(range(100))
df=pd.DataFrame({"from" : np.random.choice(node,100),"to":np.random.choice(node,100)})
g = nx.from_pandas_edgelist(df , source = "from" , target  = "to")
nx.draw(g);
```

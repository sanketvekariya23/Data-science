# z score

```python
# z score is nothing but how many standerd deviation way a datapoints is from mean.
# if i have datapoint is 2SD then z score is 2 from bell curve

# z score = X- u / SD

# x data
# u mean
# SD standerd deviation 

# zscore is also used to data cleaning for remove outlier
```

```python
import pandas as pd
df1 = pd.read_csv("hight.csv")
df1
```

```python
df1["zscore"] = (df1.Hight-df1.Hight.mean())/df1.Hight.std()
df1
```

```python
df1_no_outlier = [(df1.zscore>-3) &(df1.zscore<3)]
df1_no_outlier
```

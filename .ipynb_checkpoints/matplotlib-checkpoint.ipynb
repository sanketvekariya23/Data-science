# bar chart

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
compny = ["google" ," facebook" ," amzone" , "flpcart"]
revenue= [96,129,87,56]
profit = [56,58,60,50]
```

```python
xpos = np.arange(len(compny))
xpos
```

```python
plt.bar(xpos , revenue, label='revenue')
plt.bar(xpos,profit ,label="profit")
plt.legend()
plt.ylabel("revenue")
plt.xlabel("compny")
plt.title("compney charts")
plt.xticks(xpos,compny)
```

# horizotal bar charts


```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

compny = ["google" ," facebook" ," amzone" , "flpcart"]
revenue= [96,129,87,56]
profit = [56,58,60,50]

xpos = np.arange(len(compny))

plt.barh(xpos , revenue, label='revenue')
plt.barh(xpos,profit ,label="profit")
plt.legend(Shadow = True)
plt.ylabel("revenue")
plt.xlabel("compny")
plt.title("compney charts")
plt.yticks(xpos,compny)
```

```python
import matplotlib.pyplot as plt 

x =[10,30,30,20]
c = ['r','b','g','y']
l = ["c++" , "java" , "python" ," mysql"]
e = [0.3,0.0,0.0,0.0]
plt.pie(x,colors = c, shadow = True,labels= l,explode = e,labeldistance = 1.1,autopct="%0.2f%%",rotatelabels = True)
plt.title("pie chart")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 

x =[10,30,30,20]
y =[10,40,20,20]
c = ['r','b','g','y']
l = ["c++" , "java" , "python" ," mysql"]
e = [0.3,0.0,0.0,0.0]
plt.pie(x,colors = c, shadow = True,labels= l,explode = e,labeldistance = 1.1,autopct="%0.2f%%",rotatelabels = True)
plt.pie(y,colors = 'w',radius = 0.5)
plt.title("pie chart")
plt.legend()
plt.show()
```
